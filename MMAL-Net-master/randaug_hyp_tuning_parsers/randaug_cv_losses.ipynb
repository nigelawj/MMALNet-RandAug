```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('randaug.csv')
df.head()
```

```python
df = df.sort_values(by=['n', 'm'])
```

```python
df = df.reset_index(drop=True)
df
```

```python
x_ticks = []
for i in df.index:
    x_ticks.append(f"N={df.iloc[i]['n']:.0f} M={df.iloc[i]['m']:.0f}")
```

```python
plt.figure(dpi=1000)
plt.bar(np.arange(len(df)), height=df['avg_min_loss'], yerr=df['loss_std'])
plt.xticks(np.arange(len(df)), x_ticks, rotation='vertical')
plt.ylim([10.6, 10.9])
plt.tight_layout()
plt.savefig('randaug_cv_losses.jpg')
plt.show()
```
